* En primer lugar vamos a importar las librería necesarias:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

<hr>


### 1. Carga de Datos

```python
df = pd.read_csv('/content/drive/MyDrive/precio_casas.csv', header=0, sep=';')
df.head()
```

<hr>


### 2. Cambio de estructura de datos a numpy


```python
# Extraer las variables independientes (X) y dependientes (y)
X = df[['pies_cuadrados_casa']].values  # Variable independiente
y = df['precio_dolares'].values          # Variable dependiente
print(X.shape)
print(y.shape)
```

<hr>


### 3. Visualizar con una gráfica de dispersión la relación entre precio y pies cuadrados de las casas


```python
plt.scatter(X, y, color='blue')
plt.title('Relación entre Precio y Pies Cuadrados de la Casa')
plt.xlabel('Pies Cuadrados de la Casa')
plt.ylabel('Precio en Dólares')
plt.show()
```

<hr>


### 4. Creacción del modelo y ajuste del mismo


```python
from sklearn.linear_model import LinearRegression

# Crear el modelo
modelo = LinearRegression()

# Ajustar el modelo a los datos
modelo.fit(X, y)

```

<hr>


### 5. Obtención del modelo (Resta de Regresión)


```python
# Obtener la pendiente (coeficiente) y la intersección (intercepto)
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

print(f'Pendiente: {pendiente}')
print(f'Intercepto: {intercepto}')
```

<hr>


### 6. Visualización del modelo en el que se muestre una gráfica de dispersión con la elación entre precio y pies cuadrados de las casas y por otro lado (y en la misma gráfica) la recta de regresión.


```python
# Predicciones basadas en el modelo ajustado
y_pred = modelo.predict(X)

# Gráfica con la recta de regresión
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', label='Recta de Regresión')
plt.title('Modelo de Regresión Lineal')
plt.xlabel('Pies Cuadrados de la Casa')
plt.ylabel('Precio en Dólares')
plt.legend()
plt.show()

```

<hr>


### 7. Medición de la calidad de los resultados: Obtención del Coeficiente de Determinación R<sup>2</sup>

```python
from sklearn.metrics import r2_score

# Calcular R²
r2 = r2_score(y, y_pred)
print(f'Coeficiente de Determinación R²: {r2}')
```
